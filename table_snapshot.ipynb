```python
import tabula
import pandas as pd
import re
import os

def extract_tables_from_pdf(pdf_path, output_dir="extracted_tables"):
    """
    Extracts tables from a PDF file using tabula-py and saves them as CSV files.

    Args:
        pdf_path (str): Path to the PDF file.
        output_dir (str): Directory to save the extracted tables.
    """

    os.makedirs(output_dir, exist_ok=True)  # Create output directory if it doesn't exist

    try:
        tables = tabula.read_pdf(pdf_path, pages="all", multiple_tables=True)

        if tables:
            for i, df in enumerate(tables):
                # Data Cleaning (Example: Remove extra spaces and newlines)
                df = df.applymap(lambda x: re.sub(r'\s+', ' ', str(x)).strip())

                output_csv = os.path.join(output_dir, f"table_{i + 1}.csv")
                df.to_csv(output_csv, index=False)
                print(f"Table {i + 1} extracted and saved to {output_csv}")
        else:
            print("No tables found in the PDF.")

    except FileNotFoundError:
        print(f"Error: PDF file not found at {pdf_path}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Example usage:
pdf_file = "./pdffiles/Your-pdf-with-tables.pdf"  # Replace with your PDF file path
output_directory = "extracted_tables"
extract_tables_from_pdf(pdf_file, output_directory)

#install tabula-py
#pip install tabula-py
```
